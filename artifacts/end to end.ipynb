```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns


from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error
```

```python
##Data Gathering
```

```python
data=load_boston()
display(data)
dir(data)
```

```python
##EDA(Exploratory Data Analysis)
```

```python
df=pd.DataFrame(data.data,columns=data.feature_names)
df.head(5)
```

```python
df['Price']=data.target
```

```python
##feater engineering
```

```python
##feature selection
```

```python
##Spliting Data
```

```python
x=df.drop('Price',axis=1)
y=df['Price']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,random_state=2)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
## Model Training
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)  ##GDA best fit line
```

```python
##MOdel evaluation
```

```python
#Testing DAta

y_predict=model.predict(x_test)


mse=mean_squared_error(y_test,y_predict)
rmse=np.sqrt(mse)
mae=mean_absolute_error(y_test,y_predict)
r2_scor=r2_score(y_test,y_predict)

print(f" MSE=={mse} \n MAE=={mae}\n RMSE=={rmse}\n R2_score=={r2_scor}")
```

```python
#Training DAta

ytrain_predict=model.predict(x_train)


train_mse=mean_squared_error(y_train,ytrain_predict)
train_rmse=np.sqrt(train_mse)
train_mae=mean_absolute_error(y_train,ytrain_predict)
train_r2_scor=r2_score(y_train,ytrain_predict)

print(f" MSE=={train_mse} \n MAE=={train_mae}\n RMSE=={train_rmse}\n R2_score=={train_r2_scor}")
```
